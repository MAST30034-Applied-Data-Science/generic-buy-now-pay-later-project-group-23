### Fraud Prediction Attempt and Why We Remove All the Fraud Data
We tried to use random forest and logistic regression to predict whether a transaction will be judged as "isfraud", but the recall of both models is relatively low, indicating that the prediction effect is not good, so we decided to delete all of fraudulent transactions. It will lose a large amount of data by doing so, but as mentioned above, this also minimizes the risk, which can guarantee continuous and risk-free income in the case of the economic depression caused by the epidemic.

We also analyzed the reasons for the unsatisfactory prediction result. The first may be that the model is not good enough, and the second is that only using the probability of fraud to determine whether it is fraud may not be accurate enough. If there are more variables in addition to the probability of fraud, we can better judge fraud.

### Ranking System Building

```python
# This function helps to generate some variables
def get_stat(data_half_year):
    ## user amount 
    store_usercnt = data_half_year.groupby('name_y').user_id.nunique().reset_index()

    ## Number of product types sold
    store_merchcnt = data_half_year.groupby('name_y').merchant_abn.nunique().reset_index()

    ## Number of dates that have transactions
    store_openday = data_half_year[['name_y','order_datetime']].groupby('name_y').order_datetime.nunique().reset_index()

    ## total number of products
    store_mer_sold = data_half_year.groupby('name_y')['merchant_abn'].count()

    ## statistical data mean
    store_stat = data_half_year.groupby('name_y')['take_rate','2021','Population density 2021','Earners (persons)2017-18',
                                    'Median age of earners (years)2017-18','Sum ($)2017-18','Mean ($)2017-18','Median ($)2017-18'].mean()

    ## transaction amount
    store_gmv = data_half_year.groupby('name_y')['dollar_value'].sum()

    ## revenue level
    store_level = data_half_year.groupby('name_y')['revenue_levels'].max()
    
    ## merging
    store_df = pd.merge(store_usercnt,store_merchcnt,on = 'name_y')
    store_df = pd.merge(store_df,store_openday,on = 'name_y')
    store_df = pd.merge(store_df,store_mer_sold,on = 'name_y')
    store_df = pd.merge(store_df,store_stat,on = 'name_y')
    store_df = pd.merge(store_df,store_gmv,on = 'name_y')
    store_df = pd.merge(store_df,store_level,on = 'name_y')
    
    store_df.columns = ['store_name', 'user_cnt', 'merchant_type_num', 'open_day',
       'merchant_sold_cnt', 'avg_take_rate', 'avg_population_2021', 'avg_population_density_2021',
       'avg_earners_2017to18', 'avg_MedianAgeOfEarners2017to18',
       'avg_IncomeSum2017to18', 'avg_IncomeMean2017to18', 'avg_IncomeMedian2017to18',
       'sum_dollar_value', 'max_revenue_levels']
    
    ## income of BNPL
    store_df['revenue'] = store_df.sum_dollar_value * store_df.avg_take_rate
    
    ## derived variables
    store_df['ARPU'] = store_df.sum_dollar_value/store_df.user_cnt
    store_df['order_per_user'] = store_df.merchant_sold_cnt/store_df.user_cnt
    
    return store_df
```

```python
store_df_all = get_stat(data_half_year)
store_df_all
```

### Model Fitting
**Object：maximize gmv * take rate**

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import scipy.stats as stats
```

#### **Top 100 Among All Merchants:**

```python
results = smf.ols("revenue ~ user_cnt + open_day + merchant_sold_cnt + ARPU + order_per_user + C(max_revenue_levels)", data=store_df_all).fit()
print(results.summary())
```

```python
store_df_all['pred_revenue'] = results.predict(store_df_all)
store_df_all = store_df_all.sort_values(by = 'pred_revenue', ascending= False).reset_index(drop=1)
store_df_all.head(100)
```

## Limitation and Assumptions:
1. The external datasets are not divided by the SA2, if such datasets can be found, we can make a better geographical analysis.
2. Also, Some of the income and population data were collected before the epidemic, but all the transaction data were collect after the outbreak, this causes a timeline conflicts. In this project we assume this conflict won't affect the prediction result.
3. For those large amount of 157 dollar products, we did consider whether this is suspected of click farming, but at the suggestion of some business major friends, we decided to treat it as the most popular items in the electronic shops.
4. The ols model has fast running speed and strong interpretability, but it's not the best prediction model. If a better ranking algorithem can be built we can get a more convincing result.
5. As metioned above, we delete all the fraud transactions, this results in a lost in a large amount of data, at the same time, we cannot guarantee that other merchants will not be fraudulent in the future. So if more factors than the fraud probability can be considered, we can develop an algorithm to detect fraudulent merchants and customers.
